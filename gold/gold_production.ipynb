#####  Функция для построения гистограмм и диаграмм размаха по столбцам датасета, так чтобы диаграмма размаха располагалась под соответсвующей гистограммой
- data - датасет
- plot_cols - желаемое количество количество графиков в одной линии

```python
def hist_boxplot(data, plot_cols):
    fig, axs = plt.subplots(
        nrows=math.ceil(2*len(data.columns)/plot_cols), ncols=plot_cols,
        figsize=(20, 6*len(data.columns)), 
    )
    plt.subplots_adjust(wspace=0.2, hspace=0.3)
    i = 0
    k = 0
    while k < len(data.columns)+2:
        for col in range(plot_cols):
            if i%2 == 0:
                if  k < len(data.columns):
                    column_name = str(data.columns[k])
                    axs[i, col].hist(data[column_name], edgecolor='black')
                    axs[i, col].grid(True)
                    axs[i, col].set_title(f'{column_name}',fontsize=14, fontweight="bold")
                    axs[i, col].set_ylabel(f'Количество значений {column_name}')
                    axs[i, col].set_xlabel(f'Диапазон значений {column_name} ',fontsize=14)
                    for j in axs[i, col].patches:
                        axs[i, col].annotate('%1.2f'%(j.get_height()), (j.get_x(), j.get_height()))
            else:
                if k % plot_cols == 0:
                    k -= plot_cols
                if  k < len(data.columns):
                    column_name = str(data.columns[k])
                    if data[column_name].dtype == 'int64' or data[column_name].dtype == 'float64':
                        axs[i, col].boxplot(data[column_name],vert=False)
                        axs[i, col].grid(True)
                        axs[i, col].set_xlabel(f'Диапазон значений {column_name} ', fontsize=14)
            k += 1
        i += 1        
```

##### Фунция для вывода среднего, медианного значений и дисперсии по столбцам датасета с выбросами


```python
def data_stat(data):
    print ('Значения стобца с выбросами')
    print ('Столбец    среднее    медиана    дисперсия')
    for col in data.columns:
        print(col, round(data[col].mean(), 4), round(data[col].median(), 4), round(data[col].var(), 4))
```

##### Фунция для вывода среднего, медианного значений и дисперсии по столбцам датасета без выбросов

```python
def cleaned_data_stat(data):
    print ('Значения стобца без выбросов')
    print ('Столбец    среднее   медиана   дисперсия')
    for col in data.columns:
        min_outline = (data[col].quantile(0.25) 
                   - 1.5*(data[col].quantile(0.75) 
                          - data[col].quantile(0.25)))
        max_outline = (data[col].quantile(0.75) 
                                 + 1.5*(data[col].quantile(0.75)
                                        -data[col].quantile(0.25)))
        col_temp = data.loc[(data[col] > min_outline) &  (data[col] < max_outline)][col]
        print(col, round(col_temp.mean(), 4), round(col_temp.median(), 4), round(col_temp.var(), 4))    
        
```

##### Фунция для вывода количества выбросов  в стобце датасета

```python
def outline_count(data, column_name):
    min_outline = (data[column_name].quantile(0.25) 
               - 1.5*(data[column_name].quantile(0.75) 
                      - data[column_name].quantile(0.25)))

    max_outline = (data[column_name].quantile(0.75) 
                             + 1.5*(data[column_name].quantile(0.75)
                                    -data[column_name].quantile(0.25)))
    min_outline_count = data.loc[(data[column_name] < min_outline)][column_name].count()
    max_outline_count = data.loc[(data[column_name] > max_outline)][column_name].count()
    name = str(column_name)
    print (f'Доля выбросов в столбце {column_name} = {min_outline_count*100 / data[column_name].count()}')
    print (min_outline_count, max_outline_count, min_outline_count + max_outline_count)
```

### Определение функций для выбора модели

#### Определение функций для вычисления sMAPE

```python
def sMAPE(prediction, target):
    numerator = (abs(target - prediction))
    denominator = 0.5*((abs(target) + abs(prediction)))
    fraction = numerator / denominator
    return 100*fraction.mean()
```

#### Определение функций для вычисления взвешенной sMAPE

```python
def w_sMAPE(sMAPE_float, sMAPE_clear):
    weighted_sMAPE = 0.75*sMAPE_clear + 0.25 * sMAPE_float
    return weighted_sMAPE
```

#### Определение функций для выбора модели DecisionTreeRegressor

```python
scorer = make_scorer(sMAPE, greater_is_better=False)
```

```python
def best_model_tree(feature, target,N_cv):
    model_tree = DecisionTreeRegressor(random_state=12345)
    parameters = {'max_depth':list(range(1, 20+1))}
    clf = (GridSearchCV(
        model_tree,parameters, cv=5,scoring=scorer, return_train_score = True).
           fit(feature, target)
          )
    return clf.best_estimator_, clf.best_score_
```

#### Определение функций для выбора модели RandomForestRegressor

```python
def best_model_forest(feature, target,N_cv):
    model_tree = RandomForestRegressor(random_state=12345)
    parameters = {'max_depth':list(range(1, 10)),'n_estimators':list(range(1, 101, 10)) }
    clf = (GridSearchCV(
        model_tree,parameters, cv=5,scoring=scorer, return_train_score = True).
           fit(feature, target)
          )
    return clf.best_estimator_, clf.best_score_
```
